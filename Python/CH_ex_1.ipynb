```python
# SSAV

# import libraries
import ssav_helpers
import ssav
from ssav_setup import *
from ssav_helpers import *
from ssav import *
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fftn, ifftn, fftfreq
from dataclasses import dataclass
```

```python
L = np.array([2*np.pi, 2*np.pi])
N = np.array([128, 128])

Grid = generate_Grid(L, N, dim=2)

Time = TimeDomain(dt_min=1e-6, 
                  dt_max=0.01, 
                  t0=0, 
                  tf=8)

Para = Parameters(eps=0.1, 
                  alpha=0.0, 
                  beta=1.0,
                  delta=1e-8,
                  err_tol = np.array([1e-3, 1e-6]),
                  M=1.0, 
                  m=0.0, 
                  S=1.0, 
                  B=2.0, 
                  model="CH")
```

```python
u = 0.05 * np.sin(Grid.xx) * np.sin(Grid.yy)
```

```python
Results = ssav(Grid, Time, Para, u, dim=2)
```

```python
img = plt.imshow(Results.u, cmap = 'bwr')
img.set_clim(-1,1)

plt.colorbar
plt.show()
```
